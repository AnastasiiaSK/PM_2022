## Description of the dataset

The selected data set describes the number of passengers transported by type of transport for 1995-2020. 
It includes the following variables:
- the number of passengers transported by railway for 1995-2020.
- the number of passengers transported by sea for 1995-2020.
- the number of passengers transported by river transport for 1995-2020.
- the number of passengers transported by automobile (buses) for 1995-2020.
- the number of passengers transported by air for 1995-2020.
- the number of passengers transported by trams for 1995-2020.
- the number of passengers transported by trolleybuses for 1995-2020.
- the number of passengers transported by subway for 1995-2020.

Source: State Service of Statistics of Ukraine. Link: (http://www.ukrstat.gov.ua/operativ/menu/menu_u/tr.htm)

### Cleaning and transforming of the data set

The dataset is unclean, so we need to clean it up and convert it

```python
def transport_dataset():
    import pandas as pd
    import numpy as np
    transport = pd.read_excel('W:/University/process_mining/PM_2022/Data/kppvt_95_20_ue.xlsx', 
    skiprows=2, 
    skipfooter=4,
    usecols = "A:I",
    index_col=0,
    names=['Year', 
    'Railway', 
    'Sea', 
    'River',
    'Automobile(buses)',
    'Air',
    'Trams',
    'Trolleybuses',
    'Subway'])
    
    transport['Railway'] = transport['Railway'] * 1000
    transport['Sea'] = transport['Sea'] * 1000
    transport['River'] = transport['River'] * 1000
    transport['Automobile(buses)'] = transport['Automobile(buses)'] * 1000
    transport['Air'] = transport['Air'] * 1000
    transport['Trams'] = transport['Trams'] * 1000
    transport['Trolleybuses'] = transport['Trolleybuses'] * 1000
    transport['Subway'] = transport['Subway'] * 1000
    
    transport['Total'] = transport['Railway'] + transport['Sea'] + transport['River'] + transport['Automobile(buses)'] + transport['Air'] + transport['Trams'] + transport['Trolleybuses'] + transport['Subway']

    return transport

transport_dataset()
```

## Five question to the data set
Let's ask 5 question to the data set and answers them.

### Question 1
In which year the largest total number of passengers was transported?

```python
def answer_one():
    q_1 = transport_dataset()
    return q_1['Total'].idxmax(), q_1['Total'].max()

answer_one()
```

### Question 2
In which year was the share of passenger traffic by railway the largest?

```python
def answer_two():
    q_2 = transport_dataset()
    q_2['% Railway'] = q_2['Railway'] / q_2['Total']
    q_2['% Railway'] = q_2['% Railway'] * 100
    q_2['% Railway'] = q_2['% Railway'].round(2)
    return q_2['% Railway'].idxmax(), q_2['% Railway'].max()

answer_two()
```

### Question 3
In which year was the number of passengers carried by river more than by sea?

```python
def answer_three():
    q_3 = transport_dataset()
    q_3['Diff_River_Sea'] = q_3['River'] - q_3['Sea']
    q_3_index = q_3['Diff_River_Sea'].idxmax()
    q_3_max = q_3['Diff_River_Sea'].max()
    return q_3_index, q_3_max

answer_three()
```

### Question 4
Find min, max, mean, and std of the total number of passengers carried during the periods: 
- 1995-1999
- 2000-2004
- 2005-2009
- 2010-2014
- 2015-2020

```python
def answer_four():
    import pandas as pd
    q_4 = transport_dataset()
    q_4['Total, mln'] = q_4['Total'] / 1000000
    q_4['Total, mln'] = q_4['Total, mln'].round(2)
    YearBins  = { 1995:'1995-1999', 1996:'1995-1999', 1997:'1995-1999', 1998:'1995-1999', 1999:'1995-1999',
                2000:'2000-2004', 2001:'2000-2004', 2002:'2000-2004', 2003:'2000-2004', 2004:'2000-2004',
                2005:'2005-2009', 2006:'2005-2009', 2007:'2005-2009', 2008:'2005-2009', 2009:'2005-2009',
                2010:'2010-2014', 2011:'2010-2014', 2012:'2010-2014', 2013:'2010-2014', 2014:'2010-2014',
                2015:'2015-2020', 2016:'2015-2020', 2017:'2015-2020', 2018:'2015-2020', 2019:'2015-2020', 2020:'2015-2020' }
    
    transport_stat = pd.DataFrame(columns = ['Min, mln', 'Max, mln', 'Average, mln', 'Std, mln'])
    for group, frame in q_4.groupby(YearBins):
        transport_stat.loc[group] = [frame['Total, mln'].min(), frame['Total, mln'].max(), frame['Total, mln'].mean(), frame['Total, mln'].std()]
    return transport_stat

answer_four()
```
